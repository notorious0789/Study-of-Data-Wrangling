```python
data={
    "Roll No":[101,102,103,104,105],
      "Name":['Asha','Rohan','Meera','Karan','Anil'],
      'Age':[20,'-',21,20,'22'],
      'Marks':[85,90,'-',88,76],
      'Department':['CSE','IT','CSE','cse','E&TC'],
      'Admission_Date':['01-06-2023','03-06-2023','05-06-2023','07-06-2023','10/06/2023']
    }
df=pd.DataFrame(data)
df

df.to_csv('student_data.csv', index=False)
```

```python
#Replace '-' with NaN to mark misssing values
df.replace("-", np.nan, inplace=True)
df
```

```python
#Convert Columns to numeric where needed
df['Age']=pd.to_numeric(df['Age'], errors='coerce')
df['Marks']=pd.to_numeric(df['Marks'], errors='coerce')
df
```

```python
#Check missing values
print('\nMissing values:')
print(df.isnull().sum())
```

```python
#Fill missing age with mean
df['Age'].fillna(df['Age'].mean(), inplace=True)
df
```

```python
df['Marks'].fillna(df['Marks'].median(), inplace=True)
df
```

```python
#Fix inconsistent text formatting
df['Department']=df['Department'].str.upper()
df
```

```python
#Convert Admission_Dat to standard datetime format
df['Admission_Date']=pd.to_datetime(df['Admission_Date'],errors='coerce')
df
```

```python
print('Cleaned data')
df
```

```python
df.to_csv('cleaned_studet_data.csv', index=False)
print('Cleaned csv saed as cleaned_studet_data.csv')
```

```python
import pandas as pd

NewData = {
    "Fruits" : ["Apple","Banana","Mango","Grapes"],
    "Calories" : [420,380,"_",210],
    "Weight" : ["_",40,70,3],
    "Taste" : ["Sweet","Sweet","Sour","Sour"],
    "Price" : [70,40,"_",100],
    "DateOfPurchace" : ["2023-03-12","2020-01-06","03-01-2025","2026/02/12"]
}

df = pd.DataFrame(NewData)
print("Original DataFrame:")
df.to_csv("Fruits.csv",index=False)
print(df)
```

```python
import numpy as np

df.replace("_",np.nan,inplace=True)
print(df)
```

```python
df["Calories"]=pd.to_numeric(df["Calories"],errors="coerce")
df["Weight"]=pd.to_numeric(df["Weight"],errors="coerce")
df["Price"]=pd.to_numeric(df["Price"],errors="coerce")
print(df)
```

```python
print(df.isnull().sum())
```

```python
print(df["Calories"].fillna(df["Calories"].mean(),inplace=True))
print(df["Weight"].fillna(df["Weight"].median(),inplace=True))
print(df["Price"].fillna(df["Price"].mean(),inplace=True))
```

```python
df["Fruits"] = df["Fruits"].str.upper()
df["Taste"] = df["Taste"].str.lower()
df
```

```python
print("Cleaned data : \n",df)
df.to_csv("CleanedData.csv",index=False)
```
