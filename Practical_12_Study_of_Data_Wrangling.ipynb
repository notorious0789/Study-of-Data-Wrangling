```python
# Importing the Pandas and NumPy
import pandas as pd
import numpy as np
df = pd.DataFrame({'C1':[11,22,np.nan],'C2':[15,np.nan,np.nan],'C3':[10,20,30]})
print(df)

print(df.isna()) # Boolean Function
# isna() = is NaN(Not a Number)
# notna() : is not NaN
x = df.isna().sum()
print(x)

print(df.dropna()) # NaN rows are dropped/delete

print(df.fillna(value='_')) # NaN is replace by given value. In our case it is _

print(df.fillna(df.mean())) # Replace the NaN by mean value of the COLUMN
```
